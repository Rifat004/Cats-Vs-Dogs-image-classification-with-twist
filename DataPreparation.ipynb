```python
import pandas as pd
labels_df=pd.read_csv('/content/drive/MyDrive/Dataset/Dogs vs Cats/Dog Breed Identification/labels.csv')
labels_df
```

```python
labels_df.breed.value_counts()
```

```python
# Paths
dog_train_folder = '/content/drive/MyDrive/Dataset/Dogs vs Cats/Dog_train'
dog_test_folder = '/content/drive/MyDrive/Dataset/Dogs vs Cats/DogTest'

# Get the list of unique breeds and select 105 for training and 15 for testing
all_breeds = labels_df['breed'].unique()
breeds_for_train = all_breeds[:105]
breeds_for_test = all_breeds[105:120]

# Iterating through the labels and copy images to appropriate folders
for index, row in labels_df.iterrows():
    image_filename = f"{row['id']}.jpg"
    source_path = os.path.join('/content/drive/MyDrive/Dataset/Dogs vs Cats/Dog Breed Identification/train', image_filename)

    if row['breed'] in breeds_for_train:
        destination_path = os.path.join(dog_train_folder, image_filename)
    elif row['breed'] in breeds_for_test:
        destination_path = os.path.join(dog_test_folder, image_filename)
    else:
        continue  # Skip breeds not in selected sets

    shutil.copy(source_path, destination_path)

# Creating separate DataFrames for train and test labels (for ground truth)
train_labels_df = labels_df[labels_df['breed'].isin(breeds_for_train)]
test_labels_df = labels_df[labels_df['breed'].isin(breeds_for_test)]

train_labels_df.to_csv('/content/drive/MyDrive/Dataset/Dogs vs Cats/Dog_train_labels.csv', index=False)
test_labels_df.to_csv('/content/drive/MyDrive/Dataset/Dogs vs Cats/Dog_test_labels.csv', index=False)
```

```python
import random
import shutil

# Source directory with cat images
source_directory = '/content/drive/MyDrive/Dataset/Dogs vs Cats/Cats'

# Destination directories for training and testing
train_directory = '/content/drive/MyDrive/Dataset/Dogs vs Cats/Train/Cat'
test_directory = '/content/drive/MyDrive/Dataset/Dogs vs Cats/Test/Cat'

# Get a list of cat image filenames
cat_image_filenames = os.listdir(source_directory)
random.shuffle(cat_image_filenames)  # Shuffle the list

# Number of images for training and testing
num_train_images = 8995
num_test_images = 3505

# Move images for training
for filename in cat_image_filenames[:num_train_images]:
    source_path = os.path.join(source_directory, filename)
    destination_path = os.path.join(train_directory, filename)
    shutil.move(source_path, destination_path)

# Move images for testing
for filename in cat_image_filenames[num_train_images:num_train_images+num_test_images]:
    source_path = os.path.join(source_directory, filename)
    destination_path = os.path.join(test_directory, filename)
    shutil.move(source_path, destination_path)
```

```python
print(len(os.listdir('/content/drive/MyDrive/Dataset/Dogs vs Cats/Train/Cat')))
print(len(os.listdir('/content/drive/MyDrive/Dataset/Dogs vs Cats/Test/Cat')))
```

```python
# Source directories with dog images
dog_train_directory = '/content/drive/MyDrive/Dataset/Dogs vs Cats/Dog_train'
dog_test_directory = '/content/drive/MyDrive/Dataset/Dogs vs Cats/DogTest'

# Destination directories for training and testing
train_dog_directory = '/content/drive/MyDrive/Dataset/Dogs vs Cats/Train/Dog'
test_dog_directory = '/content/drive/MyDrive/Dataset/Dogs vs Cats/Test/Dog'

# Get a list of dog image filenames
dog_train_image_filenames = os.listdir(dog_train_directory)
dog_test_image_filenames = os.listdir(dog_test_directory)

# Move dog images for training
for filename in dog_train_image_filenames:
    source_path = os.path.join(dog_train_directory, filename)
    destination_path = os.path.join(train_dog_directory, filename)
    shutil.move(source_path, destination_path)

# Move dog images for testing
for filename in dog_test_image_filenames:
    source_path = os.path.join(dog_test_directory, filename)
    destination_path = os.path.join(test_dog_directory, filename)
    shutil.move(source_path, destination_path)
```
